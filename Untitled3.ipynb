```python
import pygame
import random
import math
pygame.init()
screen=pygame.display.set_mode((800,600))
pygame.display.set_caption("aravind")
icon=pygame.image.load('alien.png')
img=pygame.image.load('space-invaders.png')

bac=pygame.image.load('2463091.jpg')
pygame.display.set_icon(icon)
playerx=370
playery=533
playerxchange=0
enmyimg=[]
enmyx=[]
enmyy=[]
enmyxchange=[]
enmyychange=[]
num_of_enemies=6

for i in range(num_of_enemies):
    enmyimg.append(pygame.image.load('spaceship.png'))
    enmyx.append(random.randint(0,735))
    enmyy.append(random.randint(50,150))
    enmyxchange.append(2)
    enmyychange.append(40)
bulimg=pygame.image.load('bullet.png')
bulx=0
buly=533
bulychange=10
bul_state="ready"
score=0
font=pygame.font.Font('freesansbold.ttf',32)
textx=10
texty=10
over_font=pygame.font.Font('freesansbold.ttf',64)
def game_over_text():
    over=font.render("GAMEOVER"+str(score),True,(255,255,255))
    screen.blit(over,(200,250))

def shows(x,y):
    SCORE=font.render("SCORE :",True,(255,255,255))
    screen.blit(SCORE,(x,y))
    
def player(x,y):
    screen.blit(img,(x,y))
def enemy(x,y,i):
    screen.blit(enmyimg[i],(x,y))
def fire(x,y):
    global bul_state
    bul_state="fire"
    screen.blit(bulimg,(x+16,y+10))
def iscollision(enmyx,enmyy,bulx,buly):
    dist=math.sqrt((math.pow(enmyx-bulx,2))+(math.pow(enmyy-buly,2)))
    if dist<27:
        return True
    else:
        return False
    

run=True
while run:
    screen.fill((0,0,0))
    screen.blit(bac,(0,0))
   # playery-=0.1
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run=False
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_LEFT:
                playerxchange=-5
            if event.key==pygame.K_RIGHT:
                playerxchange=5
            if event.key==pygame.K_SPACE:
                bulx=playerx
                
                fire(bulx,buly);
        if event.type==pygame.KEYUP:
            if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                
                playerxchange=0
                
        
    playerx+=playerxchange
    if playerx<=0:
        playerx=0
    elif playerx>=736:
        playerx=736
    for i in range(num_of_enemies):
        if enmyy[i]>200:
            for j in range(num_of_enemies):
                enmyy[j]=2000
            game_over_text()
            break
        enmyx[i]+=enmyxchange[i]
        #enmyy+=enmyychange
    
        if enmyx[i]<=0:
            enmyxchange[i]=1
            enmyy[i]+=enmyychange[i]
        elif enmyx[i]>=736:
            enmyxchange[i]=-1
            enmyy[i]+=enmyychange[i]
        collision=iscollision(enmyx[i],enmyy[i],bulx,buly)
        if collision:
            buly=480
            bul_state="ready"
            score+=1
            enmyx[i]=random.randint(0,735)
            enmyy[i]=random.randint(50,150)  
        enemy(enmyx[i],enmyy[i],i)   
    if buly<=0:
        buly=533
        bul_state="ready"
    if bul_state is"fire":
        fire(bulx,buly)
        buly-=bulychange
    
        
    player(playerx,playery)  
    shows(textx,texty)
    
    pygame.display.update()  
```
